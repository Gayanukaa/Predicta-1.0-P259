```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/daily_data.csv')
```

```python
# Replace 'Date' with the name of your date column


# Display basic statistics
print(df.describe())
```

```python
null_counts = df.isnull().sum()
```

```python
df1 = df[df['city_id'] == 'C001']
```

```python
plt.figure(figsize=(14, 7))
plt.plot(df1.index, df1['temperature_celsius'], label='Temperature')
plt.plot(df1.index, df1['wind_kph'], label='Wind')
plt.plot(df1.index, df1['cloud'], label='Cloud')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.title('Temperature variation of city C001 Over Time')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.hist(df1['temperature_celsius'], bins=30, edgecolor='black')
plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.title('Distribution of Temperature')
plt.show()
```

```python
# Resample the data by month and calculate mean and median
monthly_mean = df1['temperature_celsius'].mean()
monthly_median = df1['temperature_celsius'].median()

print('monthly_mean',monthly_mean)
print('monthly_median',monthly_median)
```
